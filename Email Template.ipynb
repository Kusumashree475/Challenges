```python
import re
def email_body():
    phone=list()
    email=list()
    for i in range(0, df.shape[0]):
        text=df.loc[i]['Email Boby']
        #display(text)
        number=re.findall(r'\d{3,3}[-]\d{3,3}[-]\d{4,4}|[+]\d{1,3}\s+\(\d{3}\)\s+\d{3}\-\d{4}|\(\d{3}\)\s+\d{3}\-\d{4}', text)  
        if number:
            phone.append(number)
            
        email1=re.findall(r'\w+[@]\w+\.\w+',text)        
        if email1:
            email.append(email1)
       
    return phone, email
```

```python
phone, email=email_body()
```

```python
updated_phone=[]
for i in phone:
    
    for j in i:
        
        updated_phone.append(j)
display(updated_phone)


updated_email=[]
for i in email:
    for j in i:
        updated_email.append(j)

        display(updated_email)
```

```python
phone_df=pd.DataFrame()
email_df=pd.DataFrame()
#pd.DataFrame(updated_phone, columns={'phone number'})
```

```python
phone_df['Phone_number']=pd.Series(updated_phone)
email_df['Email']=pd.Series(updated_email)
```

```python
phone_df.to_csv('phone numbers.csv')
email_df.to_csv('Email list.csv')
```
